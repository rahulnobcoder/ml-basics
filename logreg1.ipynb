```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import math
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
data
```

```python
data=data.drop(columns="User ID")
```

```python
data.insert(0,"trash",1)
```

```python
X=data.drop(columns="Purchased")
Y=data["Purchased"]
X["Gender"]=X["Gender"].replace(["Male","Female"],[1,0])
Xt=X.transpose()
a=np.linalg.inv(np.matmul(Xt,X))
a=np.matmul(a,Xt)
mat=np.matmul(a,Y)
mat
```

```python
def predict(a,b,c,d):
    y=d[0]+a*d[1]+b*d[2]+c*d[3]
    return y
pred=[]
for i in range(400):
    pred.append(predict(X.iloc[i,1],X.iloc[i,2],X.iloc[i,3],mat))
pred
```

```python
for i in range(400):
    pred[i]=1/(1+math.exp(-pred[i]))
```
